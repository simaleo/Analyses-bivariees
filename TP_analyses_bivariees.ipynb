## 1. Analyse QUALITATIVE vs QUALITATIVE
#### Question 1

##### Quelle est la relation entre Zone et Pollution ?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les données
df = pd.read_excel("milieu_marin_700.xlsx")

# Tableau croisé
table = pd.crosstab(df['Zone'], df['Pollution'])
print(table)

# Graphique
table.plot(kind='bar', stacked=True)
plt.title("Zone vs Pollution")
plt.xlabel("Zone")
plt.ylabel("Effectif")
plt.show()
```

## 2. Analyse QUALITATIVE vs QUANTITATIVE
#### Question 2

##### Comment varie la Taille des poissons selon la Zone ?

```python
import matplotlib.pyplot as plt

# Boxplot
df.boxplot(column='Taille', by='Zone')

plt.title("Taille selon la zone")
plt.suptitle("")
plt.xlabel("Zone")
plt.ylabel("Taille (cm)")
plt.show()
```

#### Question 3

##### Quelle est la température moyenne selon le niveau de pollution ?

```python
# Moyenne par groupe
moyennes = df.groupby('Pollution')['Temperature'].mean()
print(moyennes)

# Graphique
moyennes.plot(kind='bar')
plt.title("Température moyenne selon pollution")
plt.ylabel("Température (°C)")
plt.show()
```

## 3. Analyse QUANTITATIVE vs QUANTITATIVE
#### Question 4

##### Existe-t-il une relation entre Température et Salinité ?

```python
plt.scatter(df['Temperature'], df['Salinite'])

plt.title("Température vs Salinité")
plt.xlabel("Température (°C)")
plt.ylabel("Salinité")
plt.show()
```

#### Question 5

##### Relation entre Taille et Poids des poissons ?

```python
plt.scatter(df['Taille'], df['Poids'])

plt.title("Taille vs Poids")
plt.xlabel("Taille (cm)")
plt.ylabel("Poids (g)")
plt.show()
```

#### Question 6 (bonus)

##### Tracer une droite de tendance (sans test statistique)

```python
import numpy as np

x = df['Taille']
y = df['Poids']

# Ajustement linéaire
coef = np.polyfit(x, y, 1)
poly = np.poly1d(coef)

plt.scatter(x, y)
plt.plot(x, poly(x))

plt.title("Relation Taille - Poids")
plt.xlabel("Taille")
plt.ylabel("Poids")
plt.show()
```

## 4. Analyse mixte (bonus pédagogique)
#### Question 7

##### Comparer le poids moyen selon les espèces

```python
moyennes = df.groupby('Espece')['Poids'].mean()
print(moyennes)

moyennes.plot(kind='bar')
plt.title("Poids moyen par espèce")
plt.show()
```
